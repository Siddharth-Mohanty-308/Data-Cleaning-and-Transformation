### 1. Data type of height and weight columns changed to integer

```python
df_3.dtypes
```

```python
df_3['Height'].unique()  # We need to handle both "170cm" and 6\'5" 
```

```python
#Function to convert

def convert_height_to_int(height_str):
    if 'c' in height_str:
        new_list=height_str.split('c')
        return int(new_list[0])
    else:
        new_list=height_str.split("'")
        total_inches=(int(new_list[0])*12) + (int(new_list[1][0]))
        new_height_cm=total_inches*2.54
        return round(new_height_cm)   
```

```python
df_4=df_3.copy()
df_4['Height']=df_3['Height'].apply(convert_height_to_int)
df_4
```

```python
df_4["Height"].unique()
```

```python
#Changing name of column for readability
#df_4.rename(columns={'Height':'Height(cm)'},inplace=True)
```

```python
df_4["Weight"].unique() # We need to handle both kg and lbs values
```

```python
#Function for weight

def convert_weight_to_int(weight_str):
    if 'k' in weight_str:
        new_list=weight_str.split('k')
        return int(new_list[0])
    else:
        new_list=weight_str.split('l')
        weight_in_kg=int(new_list[0])*0.454
        return round(weight_in_kg)   
```

```python
df_5=df_4.copy()
df_5['Weight']=df_4['Weight'].apply(convert_weight_to_int)
df_5.head()
```

```python
df_5["Weight"].unique()
```

```python
df_5.head()
```

```python
# Changing name of colunm 
df_5.rename(columns={'Height':'Height(cm)','Weight':'Weight(kg)'},inplace=True)
df_5.head()
```

### 2. Join column separated into join_year, join_month and join_day column

```python
df_5.head()
```

```python
df_5['Joined']=pd.to_datetime(df_5['Joined'])
df_5['Join_day']=df_5['Joined'].dt.day
df_5['Join_month']=df_5['Joined'].dt.month
df_5['Join_year']=df_5['Joined'].dt.year

df_6=df_5.drop('Joined',axis=1)
df_6.head()
```

### 3. Transform value,wage,release clause columns into integer

```python
# We find that all currencies are in euros
df_6["Wage"].unique()
    
```

```python
def convert_wage_to_int(wage_str):
    new_str=wage_str.lower().replace('€','').replace('k','000')
    return int(new_str)
```

```python
# We need to convert both M and K to integers
# There is a 0 euro value 
df_6['Value'].unique()
```

```python
# Drop 0 values from Values and Release Clause columns
df_6=df_6[df_6['Value'] != '€0']
df_6=df_6[df_6['Release Clause'] != '€0']
```

```python
def convert_value_to_int(value_str):
    if 'k' in value_str.lower():
        new_str=value_str.lower().replace('€','').replace('k','000')
        return int(new_str)
    elif 'm' in value_str.lower():
        new_str=value_str.lower().replace('€','').replace('m','')
        final_int=float(new_str)*1000000
        return int(final_int)
    else:
        print(value_str)
        return int(value_str)
```

```python
# We can use same function for release clause
print(df_6['Release Clause'].unique())
```

```python
df_7=df_6.copy()
df_7['Wage']=df_6['Wage'].apply(convert_wage_to_int)
df_7['Value']=df_6['Value'].apply(convert_value_to_int)
df_7['Release Clause']=df_6['Release Clause'].apply(convert_value_to_int)
df_7.head()
```

```python
# Changing names of columns
df_7.rename(columns={'Value':'Value(€)','Wage':'Wage(€)','Release Clause':'Release Clause(€)'},inplace=True)
df_7.head()
```

### 4. Remove new line charecter from club column and clean hits column

```python
def transform_club_column(club_str):
    return club_str.replace('\n','')
```

```python
df_7['Hits'].unique()
```

```python
def transform_hits_column(hits_value):
    if 'k' in str(hits_value).lower():
        new_str=hits_value.lower().replace('k','')
        final_str=float(new_str)*1000
        return int(final_str)
    else:
        return int(hits_value)
```

```python
df_8=df_7.copy()
df_8['Club']=df_7['Club'].apply(transform_club_column)
df_8['Hits']=df_7['Hits'].apply(transform_hits_column)
df_8.head()
```
